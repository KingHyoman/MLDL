```python
Auto_column = Auto.columns

for column in Auto_column:
    if column == 'name':
        continue
    max, min = Auto[column][0], Auto[column][0]
    for i in Auto[column]:
        if i == '?':
            continue
        if float(i) > float(max):
            max = float(i)
        if float(i) < float(min):
            min = float(i)
    print(f'the range of {column} is {max - min}')
    print(f'[{max}, {min}]')
```

```python
import matplotlib.pyplot as plt

Auto = Auto_backup
#Auto

arr = []

for column in Auto_column:
    if column == 'name':
        continue
    tem = (column, [])
    for i in Auto[column]:
        if i == '?':
            i = 0
        tem[1].append(float(i))
    arr.append(tem)

for i in range(len(arr)):
    for j in range(len(arr)):
        x = arr[i][1]
        y = arr[j][1]

        if len(x) != len(y):
            continue
        plt.scatter(x, y)
        plt.xlabel(arr[i][0])
        plt.ylabel(arr[j][0])
        plt.show()
```

1. There are some linear relationship: displacement - weight, horespower - weight.
2. Origin, year and cylinders have constant relationship with the others.
